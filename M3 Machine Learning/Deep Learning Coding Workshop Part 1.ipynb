## Define your model.

Hint: Check out [torch.matmul()](https://pytorch.org/docs/stable/generated/torch.matmul.html) to do a matrix multiplication

```python
class MatrixLayer(nn.Module):
  def __str__(self):
    return f"MatrixLayer({self.input_size}, {self.output_size})"

  def __init__(self, input_size, output_size):
    super(MatrixLayer, self).__init__()
    self.input_size = input_size
    self.output_size = output_size
    self.matrix = nn.Parameter(torch.randn(self.input_size, self.output_size))
    self.bias = nn.Parameter(torch.randn(self.output_size))

  def forward(self, x):
    # out = None
    # ================================= #
    #          Your code here           #
    # ================================= #
    out = torch.matmul(x, self.matrix) + self.bias
    # ================================= #

    return out
```

```python
torch.manual_seed(0)

# Initialize and visualize a 3 x 3 matrix layer
layer = MatrixLayer(3, 3)
print(layer)
print("Weights", layer.matrix)

# Create some random matrix of size 2 x 3
x = torch.randn(2, 3)
print("Data matrix", x)

# Perform a forward pass
##this is the same thing as layer.forward(x)!
out = layer(x)

print("Result", out)
```
